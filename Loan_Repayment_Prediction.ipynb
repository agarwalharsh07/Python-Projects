```python
sns.set_style('darkgrid')
plt.hist(loan['fico'].loc[loan['credit.policy']==1], bins=30, label='Credit.Policy=1')
plt.hist(loan['fico'].loc[loan['credit.policy']==0], bins=30, label='Credit.Policy=0')
plt.legend()
plt.xlabel('FICO')
```

```python
plt.figure(figsize=(10,6))
loan[loan['not.fully.paid']==1]['fico'].hist(bins=30, alpha=0.5, color='blue', label='not.fully.paid=1')
loan[loan['not.fully.paid']==0]['fico'].hist(bins=30, alpha=0.5, color='green', label='not.fully.paid=0')
plt.legend()
plt.xlabel('FICO')
```

```python
#creating a countplot to see the counts of purpose of loans by not.fully.paid
plt.figure(figsize=(12,6))
sns.countplot(data=loan, x='purpose', hue='not.fully.paid')
```

```python
#checking the trend between FICO and the interest rate
plt.figure(figsize=(10,6))
sns.jointplot(x='fico', y='int.rate', data=loan)
```

```python
#understanding the relationship between credit.policy and not.fully.paid
sns.lmplot(data=loan, x='fico', y='int.rate', hue='credit.policy', col='not.fully.paid', palette='Set2')
```
